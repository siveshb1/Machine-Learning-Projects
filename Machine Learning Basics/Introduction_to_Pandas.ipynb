```python
import pandas as pd
```

```python
car_sales = pd.read_csv("/content/004 car-sales.csv")
```

```python
car_sales_missing = pd.read_csv("/content/009 car-sales-missing-data.csv")
```

```python
car_sales_missing["Price"] = car_sales_missing["Price"].replace('[\,\$]','',regex = True)
```

```python
car_sales_missing["Odometer"] = car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean())
```

```python
car_sales_missing_dropped = car_sales_missing.dropna()
```

```python
car_sales["Price"].replace('[\$\,]','',regex = True, inplace = True)
```

```python
car_sales["Price"] = car_sales["Price"].astype(float)
```

```python
car_sales
```

```python
car_sales["Price"].dtype
```

```python
seats = pd.Series([5,5,5,5,5])
car_sales['Seats'] = seats
```

```python
car_sales
```

```python
car_sales['Seats'].fillna(5, inplace=True)
```

```python
car_sales
```

```python
fuels = [5,5.5,6,5,7,4,5,7,6.5,5.7]
car_sales["Fuel per 100km"] = fuels
```

```python
car_sales
```

```python
car_sales["Total Fuel"] = car_sales['Odometer (KM)']/100 * car_sales["Fuel per 100km"]
car_sales
```

```python
car_sales['Number of Wheels'] = 4
car_sales['Safety'] = True
```

```python
car_sales
```

```python
car_sales = car_sales.drop('Total Fuel',axis = 1)
```
