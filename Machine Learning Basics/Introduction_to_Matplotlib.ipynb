```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame
data = {'A': [1, 2, 3, 4, 5], 'B': [6, 5, 8, 2, 7], 'C': [4, 1, 5, 8, 2]}
df = pd.DataFrame(data)

# Create a figure with 2 rows and 1 column of subplots
fig, axes = plt.subplots(2, 1, figsize=(10, 6))

# Plot columns A, B, and C on separate subplots
df.plot(kind='line', x='index', y='A', ax=axes[0])  # Plot A on first subplot
df.plot(kind='line', x='index', y='B', ax=axes[1])  # Plot B on second subplot
df.plot(kind='line', x='index', y='C', ax=axes[1])  # Plot C on second subplot (shares with B)

# Customize the plot (optional)
axes[0].set_title('Column A')
axes[1].set_title('Columns B & C')
plt.xlabel('Index')

# Display the plot
plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```

```python
over_50 = heart_disease[heart_disease['age'] >50]
over_50.head()
```

```python
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target');
```

```python
over_50.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter', x='age', y='chol', c='target', ax=ax)

ax.set_xlim([45,100]);
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'over_50' is a pandas DataFrame
fig, ax = plt.subplots(figsize=(10, 6))  # Create figure and axis

# Create the scatter plot with coloring based on 'target'
scatter = ax.scatter(x='age', y='chol', c='target',data=over_50)  # Use 'ax' and 'scatter' variable

# Create a legend for the scatter plot based on 'target' values
legend = ax.legend(*scatter.legend_elements(), title='Target')  # Unpack legend elements and set title

# Optional: Customize the plot (title, labels, etc.)
ax.set_title('Cholestrol vs Age (Colored by Target)')
ax.set_xlabel('Age')
ax.set_ylabel('Cholesterol')

plt.show()
```

```python
fig,ax = plt.subplots(2,2)
heart_disease['age'].plot(kind='hist',ax=ax[0,0])
ax[0,0].set_title('Age Distribution')
ax[0, 0].set_xlabel('Age')
plt.tight_layout()
```

```python
fig,ax = plt.subplots(2,1,figsize=(10,10),sharex=True)
scatter = ax[0].scatter(x='age', y='chol', c='target',data=over_50)
ax[0].set(title='Heart Disease and Cholesterol Levels', ylabel = 'Cholesterol')
ax[0].legend(*scatter.legend_elements(),title='Target')
ax[0].axhline(y=over_50['chol'].mean(),linestyle='--');

scatter1 = ax[1].scatter(x='age',y='thalach',c='target',data=over_50)
ax[1].set(title='Heart Disease and Max Heart Rate', xlabel='Age', ylabel='Max Heart Rate')
ax[1].legend(*scatter1.legend_elements(),title='Target')
ax[1].axhline(y=over_50['thalach'].mean(),linestyle='--')

fig.suptitle('Heart Diease Analysis',fontsize=16,fontweight='bold');
```

```python
fig,ax = plt.subplots(2,1,figsize=(10,10),sharex=True)
scatter = ax[0].scatter(x='age', y='chol', c='target',cmap='winter', data=over_50)
ax[0].set(title='Heart Disease and Cholesterol Levels', ylabel = 'Cholesterol')
ax[0].set_xlim([50,80])
ax[0].legend(*scatter.legend_elements(),title='Target')
ax[0].axhline(y=over_50['chol'].mean(),linestyle='--',color='red');

scatter1 = ax[1].scatter(x='age',y='thalach',c='target',cmap='winter',data=over_50)
ax[1].set(title='Heart Disease and Max Heart Rate', xlabel='Age', ylabel='Max Heart Rate')
ax[1].set_xlim([50,80])
ax[1].set_ylim([60,200])
ax[1].legend(*scatter1.legend_elements(),title='Target')
ax[1].axhline(y=over_50['thalach'].mean(),linestyle='--',color='red')

fig.suptitle('Heart Diease Analysis',fontsize=16,fontweight='bold');
```
